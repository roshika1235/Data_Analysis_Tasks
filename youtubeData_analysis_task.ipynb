## Importing all necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading youtube_analysis dataset

```python
df=pd.read_csv("https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
df
```

## Handling null values


```python
df.info()
```

```python
df["Country"].isnull().sum()
```

```python
df["Ranking"].unique()
mean=df['Ranking'].mean()
mean=round(mean)
df["Ranking"]=df["Ranking"].fillna(mean)
df["Ranking"]=df["Ranking"].astype(int)
df["Ranking"].unique()
```

```python
df["Username"].unique()
df["Username"]=df["Username"].fillna("other channel")
```

```python
df["Subscribers"].unique()
df["Subscribers"]=df["Subscribers"].str.strip(" M")
df["Subscribers"]=df["Subscribers"].astype(float)
df["Subscribers"]=df["Subscribers"] * 1000000
median=df["Subscribers"].median()
df["Subscribers"]=df["Subscribers"].fillna(median)
df["Subscribers"]=df["Subscribers"].astype(int)
df.info()
```

```python
df["Uploads"].unique()
df["Uploads"]=df["Uploads"].fillna("0")
df["Uploads"]=df["Uploads"].str.replace(",","")
df["Uploads"]=df["Uploads"].astype(int)
df.info()
```

```python
df["Views"].unique()
df['Views']=df['Views'].str.replace(",","")
df["Views"]=df["Views"].fillna("0")
df["Views"]=df["Views"].astype(float)
df["Views"]=df["Views"].astype(int)
df.info()
```

```python
df["Country"].unique()
mode=df["Country"].mode()[0]
df["Country"]=df["Country"].fillna(mode)
df.info()
```

## EDA PROCESS

```python
plt.figure(figsize=(20,8))
plt.xticks(rotation=90)
plt.bar(df["Username"],df["Views"])
```

## people show zero interest on for Youtube Movies, sports,Gaming ,other channel.

```python
df_group = df.groupby("Country")["Views"].sum()
plt.figure(figsize=(15, 7))
plt.bar(df_group.index, df_group.values)
plt.xlabel("Country")
plt.ylabel("Total Views")
plt.title("Total Views by Country")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## india has highest views followed by usa

```python
## TOP 5 COUNTRIES WITH HIGH SUBSCRIBERS
df_top5=df.groupby("Country")["Subscribers"].value_counts().head(5)
df_top5
plt.pie(df_top5.values,labels=df_top5.index,autopct="%.2f%%")
plt.show()
```

## only 2 countries [AE,AR] have more subscribers, out of which top 3 channels with highest subscribers belong to country [United Arab Emirates]
