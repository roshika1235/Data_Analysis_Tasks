## import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## load dataset and explore


```python
df = pd.read_csv("/content/restaurant_data.csv")
```

```python
df.head()
```

## removing unwanted columns and duplicates

```python
df.drop('Name',axis=1,inplace=True)
```

```python
df.shape
```

## identifying null values

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Location'] = le.fit_transform(df['Location'])
df['Cuisine'] = le.fit_transform(df['Cuisine'])
df['Parking Availability'] = le.fit_transform(df['Parking Availability'])
```

## EDA process, data visulaization

```python
sns.barplot(x='Parking Availability',y='Revenue',data=df, hue='Location')
plt.show()
```

```python
sns.barplot(x='Location',y='Cuisine',data=df)
plt.show()
```

```python
sns.histplot(x='Number of Reviews',data=df, hue='Location')
plt.show()
```

```python
sns.heatmap(df.corr(),annot=False, cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.show()
```
